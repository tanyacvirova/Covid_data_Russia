```python
import pandas as pd
```

```python
# mining data all over the world
corona_data = pd.read_excel('https://www.ecdc.europa.eu/sites/default/files/documents/COVID-19-geographic-disbtribution-worldwide.xlsx')
corona = pd.DataFrame(corona_data)
```

```python
# select data only for Russia
corona_Russia = corona[corona['countriesAndTerritories'].str.contains('Russia')].reset_index()
corona_Russia = corona_Russia.drop('popData2019', 1).drop('continentExp', 1).drop('index', 1).drop('countryterritoryCode', 1).drop('geoId', 1)
corona_Russia = corona_Russia.drop('countriesAndTerritories', 1)
corona_Russia
```
